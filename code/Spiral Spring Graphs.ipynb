```python
import matplotlib.pyplot as plt
import numpy as np

# Given reaction force data
reaction_force = [
    9.9835e-17, 5.0874, 10.485, 16.424, 23.538, 31.782,
    41.201
]

# Generate displacement values (0, 0.5, 1.0, ..., step of 0.5)
displacement = np.arange(0, len(reaction_force) * 1, 1)

# Linear regression (fit a line to the data)
slope, intercept = np.polyfit(displacement, reaction_force, 1)
regression_line = slope * displacement + intercept

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(displacement, reaction_force, marker='o', linestyle='-', color='b', markersize=5, label="Reaction Force")
plt.plot(displacement, regression_line, color='r', linestyle='--', label="Linear Regression Line")

# Formatting
plt.xlabel("Displacement (mm)", fontsize=24)
plt.ylabel("Reaction Force (N)", fontsize=24)
plt.grid(True, linestyle="--", alpha=0.7)
plt.tight_layout()
plt.tick_params(axis='both', labelsize=24)

equation_text = f"y = {slope:.4f}x + {intercept:.2f}"
print(equation_text)
#plt.text(0.5, max(reaction_force) * 0.8, equation_text, fontsize=16, color="red")

# Show the graph
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
amplitude = np.array([0, 1, 2, 3, 4, 5, 6])  # mm
reaction_force = np.array([0, 16.67, 33.33, 50.62, 68.64, 87.58, 107.1])  # N

# Perform linear regression (1st-degree polynomial fit)
coeffs = np.polyfit(amplitude, reaction_force, 1)  # Linear fit (degree=1)
linear_fit = np.poly1d(coeffs)

# Generate fitted values
amplitude_fit = np.linspace(min(amplitude), max(amplitude), 100)
reaction_force_fit = linear_fit(amplitude_fit)

# Plot
plt.figure(figsize=(8, 6))
plt.plot(amplitude, reaction_force, marker='o', linestyle='-', markersize=8, linewidth=2)
plt.plot(amplitude_fit, reaction_force_fit, linestyle='--', color='red', linewidth=2)

# Equation of linear regression
equation_text = f"y = {coeffs[0]:.2f}x + {coeffs[1]:.2f}"
print(equation_text)

# Labels and Formatting
plt.xlabel('Amplitude (mm)', fontsize=30)
plt.ylabel('Max Stress (MPa)', fontsize=30)
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.grid(True)

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

values = [42, 41, 43, 41]
categories = ["B 1%", "B 2.5%", "B 5%", "MHEC+CSP"]

# Conversion factor
sq_in_to_sq_m = 0.0254**2  # 1 in = 0.0254 m, so square inches to square meters

# Area of each sample in square meters
area_sq_m = 5 * 3.5 * sq_in_to_sq_m

# Convert weights from grams to kg/m
values_kg_m2 = [(weight / 1000) / area_sq_m for weight in values]

# Create the updated chart
plt.figure(figsize=(8, 5))
bars = plt.bar(categories, values_kg_m2, color=['#3498DB', '#3498DB', '#3498DB', '#8E44AD'])

# Add values in the middle of each column
for bar, value in zip(bars, values_kg_m2):
    plt.text(bar.get_x() + bar.get_width() / 2,
             bar.get_height() / 2,
             f'{value:.2f}',
             ha='center',
             va='center',
             fontsize=24,
             color='white')

# Add labels and title with specific font sizes
plt.ylabel('Mass Density (kg/m)', fontsize=22)
plt.title('Mass Density Comparison', fontsize=22)
plt.xticks(fontsize=22)  # Set font size for x-axis ticks
plt.yticks(fontsize=22)  # Set font size for y-axis ticks

plt.axhline(y=3.2, color='red', linewidth=3)

plt.show()
```
